```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("DataforSLR.csv")
df = df.dropna()
df
```

```python
%matplotlib inline

plt.xlabel('area (sq ft)')
plt.ylabel('price (AU$)')

plt.scatter(df.area,df.price,color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
predicted_price = reg.predict(pd.DataFrame({'area': [3300]}))  #chatgpt
print(f"Predicted price for 3300 sq ft: {predicted_price[0]:,.2f} AU$") #chatgpt
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
data_area = pd.read_csv('ListofAreas.csv')
data_area.head(3)
```

```python
prices = reg.predict(data_area)
data_area['prices'] = prices
data_area
```

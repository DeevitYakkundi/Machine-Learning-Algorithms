```python
#find income for year 2020
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("canada_per_capita_income.csv")
df = df.dropna()
df.head(5)
```

```python
%matplotlib inline

plt.xlabel('year')
plt.ylabel('per capita income(US$)')

plt.scatter(df.year,df.per_capita_income,color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
```

```python
predicted_per_capita_income = reg.predict(pd.DataFrame({'year': [2020]}))  
print(f"Predicted price for 3300 sq ft: {predicted_per_capita_income[0]:,.2f} US$")
```
